```python
n = 50
N = 1000
tabVectorZ = []

for i in range(N):
    tab = npr.uniform(0,1,n)
    tabVectorZ.append((sum(tab) - n/2) / math.sqrt(n/12))
    
ax = np.linspace(3,3,25)
```

```python
sb.histplot(tabVectorZ,stat='density')
```

```python
tabMarmotte = []

def marmotte(n):
    for i in range(n):
        tabMarmotte.append(npr.exponential(1/0.5))
    return sum(tabMarmotte)/n
    
print("Espérance expérience:",marmotte(1000000))
print("Espérance: ",1/0.5)
print("Variance: ",1/(0.5**2))
```
